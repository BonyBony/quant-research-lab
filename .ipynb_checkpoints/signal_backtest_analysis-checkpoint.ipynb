```python
# signal_backtest_analysis.ipynb

# 🧪 Load Modules
from strategies.rsi_signal import rsi_signal
from backtest.engine import vectorized_long_only_backtest
from backtest.metrics import compute_all_metrics
from utils.visualization import plot_heatmap
from utils.data_loader import load_data

import pandas as pd

# 🔁 Parameter Grid
rsi_vals = [25, 30, 35]
hold_vals = [3, 5, 10]
results = []

# 🔄 Strategy Testing Loop
for rsi_thresh in rsi_vals:
    for hold_days in hold_vals:
        df = load_data('SPY')
        df = rsi_signal(df, rsi_thresh=rsi_thresh)
        df = vectorized_long_only_backtest(df, hold_days=hold_days)
        metrics = compute_all_metrics(df)
        results.append({
            'RSI_Threshold': rsi_thresh,
            'Holding_Days': hold_days,
            **metrics
        })

# 📊 Metrics Table
metrics_df = pd.DataFrame(results)
metrics_df.sort_values(by='Sharpe Ratio', ascending=False, inplace=True)
metrics_df.reset_index(drop=True, inplace=True)
metrics_df.head()

# 🔥 Heatmap Visualization
plot_heatmap(metrics_df, row='RSI_Threshold', col='Holding_Days', value='Sharpe Ratio')
```
